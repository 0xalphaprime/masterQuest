__DNS__

If instead of using the `ping` command, we actually want to interact with a website, we introduce a new layer. 

The DNS Server will maintain an IP Address on the LAN. 

1. Node sends message out, _"Hey, who owns this IP?"_
2. DNS Server -> _"Hey, that's me. Here is my MAC Address"_
3. Node: _"Sweet, where is this? <www.networkchuck.coffee>"_
4. DNS: _"No problem dude, here you go!"_
5. Node sends a message to the switch, _"Can you connect me with the gateway, please? <http.get() request>"_
6. Router is a layer 3 device, so it can handle IP Addresses, but it doesnt know how to tell the switch the L2 address.
7. Router does its own ARP message with IP Address to get the MAC address of the intended server. 
8. <www.networkchuck.coffee> responds with their MAC address. 
9. The switch to the coffee server responds to the router and back to the requesting node. 
10. Node now has the L2 & L3 information it needs to get that coffee!

__Encapsulation__ & __Headers__ are key themes to master. 

Below runs through a basic example of networking protocol woriking with a simle web request:

1. Application layer generates data, using the HTTPS protocol (in this instance).
2. Data is __ENCAPSULATED__ with a L4 Header (referred to as a __SEGMENT__)
    - TCP or UDP - TCP is more reliable, while UDP is known to maximize for speed
    - Ports 
3. The __SEGMENT__ is then __ENCAPSULATED__ with a L3 Header (referred to as a __PACKET__)
    - Source IP and Destination IP
4. The __PACKET__ is then __ENCAPSULATED__ with both a L2 Header & a L2 Trailer (referred to as a __FRAME__) 
5. At this point the switch knows where to send the packet, and the transport process can begin. 

![tcp](/Users/hiro0x/Documents/alphaprime/networking/screenshots/tcp.png)


Expanding this example to look  more into the specifics of what is happening during the encapsulation process. 

- Layer 6 - __PRESENTATION__ data format (file types - .html, .pdf, .docx, .png...)
- Layer 5 - __SESSION__ open up the communication and keeps it open, and tears it down when the session is over (L2TP, RTCP, SOCKS, etc. protocols at the session layer)

All of this is happening at the APP layer looking at the __TCP/IP__

- Layer 4 - __TRANSPORT__ here we select a transport protocol __TCP__ or __UDP__
    - __TCP__ Transmission Control Protocol - reliable - requires a response that you received the data, and then proceeds with the next packet(s) or resends the previous
        - 3-Way handshake [SYN] [SYN-ACK] [ACK]
    - __UDP__ User Datagram Protocol - maximized for speed - does not wait for a response from the destination, and instead just keeps sending

    - __PORTS__ analogous to a phone extension. a port is a landing place for a direct connection to a network. HTTPS, for example is accessed through port 443
        - This way we can run multiple services on one server. 
        - Important to note, that this connection is a two way street. the "from" used for the web server to connect back with you will be a non-normal port # (i.e. 57095)
            - known as __EPHEMERAL__, unique to a particular connection instance - now I can have youtube, gmail, rpl, etc. all communication back to my server.

### FIBER OPTICS

__FIBER OPTICS__ uses light vs electircal signal. FIber has low __ATTENUATION__ which is the rate the signal degrades. An issue with copper (ethernet).

Fiber also has no EMI, or cross-talk. 

Refraction, due to chaning the medium the light is going through, and can be conditioned to find the 'critical angle' for the light to travel.

Single-Mode, and Multi-Mode are both options. core in a single is a lot smaller (3-9 microns vs 125 microns in multi-mode). Single has less room to bounce aroound. Lower attenuation, higher bandwidhth... but we use both. Single is more expensive, due to engineering difficulty involving the smaller core. 

Single-Mode is most often glass, multi-mode more evenly spread across glass/plastic

fiber-optic connectors:

- ST
- SC
- LC
- MT-RJ (mini RJ45)

Characteristics & Parameters {
    **MODE:** Single-Mode, Multi-Mode
    **CONNECTOR:** ST, SC, LC, MT-RJ
    **Core Size:** small (single-mode) -> large (multi-mode)
    **Color:** corresponds to the mode. 
    **Bandwidth:** OM1(100Mb/s), OM2(1Gb/s), OM3(10Gb/s), OM4(40-100Gb/s) OM5
    **Cable Structure:** Outer Jacket => Strength Member => Coating => Cladding => core


Accessible on traditonal ethernet switches by using the SFP ports, with the apporpriate connector. SFP is the wild-card port(s).
